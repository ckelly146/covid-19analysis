```python
#London VISUALISING THE DATA
import pandas as pd
import matplotlib.pyplot as plt
import json
# an iPython  "magic" that enables the embedding of matplotlib output
%matplotlib inline
# make figures larger
plt.rcParams['figure.dpi'] = 100

with open("London.json", "rt") as INFILE:
    data=json.load(INFILE)
data
```

```python
datalist=data['data']
datalist
```

```python
dates=[dictionary['date'] for dictionary in datalist ]
dates.sort()
dates
```

```python
def parse_date(datestring):
    """ Convert a date string into a pandas datetime object """
    return pd.to_datetime(datestring, format="%Y-%m-%d")
```

```python
startdate=parse_date(dates[0])
enddate=parse_date(dates[-1])
print (startdate, ' to ', enddate)
```

```python
index=pd.date_range(startdate, enddate, freq='D')
Londondf=pd.DataFrame(index=index, columns=['Cases', 'Ventilators', 'Deaths'])
Londondf
```

```python
for entry in datalist: # each entry is a dictionary with date, cases, hospital and deaths
    date=parse_date(entry['date'])
    for column in ['Cases', 'Ventilators', 'Deaths']:
        # check that nothing is there yet - just in case some dates are duplicated,
        # maybe with data for different columns in each entry
        if pd.isna(Londondf.loc[date, column]): 
            # replace None with 0 in our data 
            value= float(entry[column]) if entry[column]!=None else 0.0
            # this is the way you access a specific location in the dataframe - use .loc
            # and put index,column in a single set of [ ]
            Londondf.loc[date, column]=value
            
# fill in any remaining "holes" due to missing dates
Londondf.fillna(0.0, inplace=True)
            
Londondf
```

```python
import seaborn as sns
sns.set_style("dark")
Londondf.plot()
plt.title("London Covid-19 Cases and Deaths 2020")
plt.ylabel("Cases")
```

```python
import seaborn as sns
sns.set_style("dark")
Londondf.plot(logy=True)
plt.title("London Covid-19 Cases and Deaths 2020")
plt.ylabel("Cases in Log")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# an iPython  "magic" that enables the embedding of matplotlib output
%matplotlib inline
# make figures larger
plt.rcParams['figure.dpi'] = 100
```

```python
with open("LDNage.json", "rt") as INFILE:
    data=json.load(INFILE)
```

```python
datadic=data['data'][0] # data['data'] is a list
datadic
```

```python
datalist=data['data']
datalist
```

```python
Males=datadic['Males']
Females=datadic['Females']
Males # females is similar 
```

```python
ageranges=[x['age'] for x in Males] # each entry of males is a dictionary
print(ageranges)
```

```python
def min_age(agerange):
    agerange=agerange.replace('+','') # remove the + from 90+
    start=agerange.split('_')[0]
    return int(start)

print(min_age('85_to_89'))
print(min_age('90+')) # looking good...
```

```python
ageranges.sort(key=min_age)
print (ageranges)
```

```python
ageLDN_df=pd.DataFrame(index=ageranges, columns=['Males','Females', 'Total'])
```

```python
for entry in Males: # each entry is a dictionary
    ageband=entry['age'] # our index position
    ageLDN_df.loc[ageband, 'Males']=entry['value']
    
for entry in Females:
    ageband=entry['age']
    ageLDN_df.loc[ageband, 'Females']=entry['value']
    
# this is straightforward
ageLDN_df['Total']=ageLDN_df['Males']+ageLDN_df['Females']
```

```python
ageLDN_df
```

```python
import seaborn as sns
sns.set_style("dark")
ageLDN_df.plot(kind='bar', y=['Males','Females','Total'])
plt.title("London Covid-19 Cases 2020 by Age and Gender")
plt.xlabel("Age")
plt.ylabel("Cases")
```

```python
import seaborn as sns
sns.set_style("dark")
ageLDN_df.plot(kind='bar', y='Total')
plt.title("London Covid-19 Cases 2020 by Age")
plt.xlabel("Age")
plt.ylabel("Cases")
```

```python
MALETOTALDN = sum(ageLDN_df['Males'])
print(MALETOTALDN)
FEMALETOTALDN =sum(ageLDN_df['Females'])
print(FEMALETOTALDN)
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Males', 'Females'
sizes = [MALETOTALDN,FEMALETOTALDN]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("London Covid-19 Cases 2020 by Gender")
plt.show()
```
